```python
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
colors = ["#69A2B0", "#9B6BAE", "#E8A87C"]
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

ax.plot([1, 9], [0.64, 0.78], color=colors[0], marker='o', linewidth=5, markersize=12)
ax.annotate('64%', xy=(1, 0.64), xytext=(-1.5, 0.64), fontsize=40, color=colors[0], va='center')
ax.annotate('78%', xy=(9, 0.78), xytext=(9.3, 0.78), fontsize=40, color=colors[0], va='center')
ax.annotate('Packages on time', xy=(9, 0.78), xytext=(12, 0.78), fontsize=40, color=colors[0], va='center')

ax.plot([1, 9], [0.56, 0.47], color=colors[1], marker='o', linewidth=5, markersize=12)
ax.annotate('56%', xy=(1, 0.56), xytext=(-1.5, 0.56), fontsize=40, color=colors[1], va='center')
ax.annotate('47%', xy=(9, 0.47), xytext=(9.3, 0.47), fontsize=40, color=colors[1], va='center')
ax.annotate('Vehicle downtime', xy=(9, 0.47), xytext=(12, 0.47), fontsize=40, color=colors[1], va='center')

ax.plot([1, 9], [0.22, 0.33], color=colors[2], marker='o', linewidth=5, markersize=12)
ax.annotate('22%', xy=(1, 0.22), xytext=(-1.5, 0.22), fontsize=40, color=colors[2], va='center')
ax.annotate('33%', xy=(9, 0.33), xytext=(9.3, 0.33), fontsize=40, color=colors[2], va='center')
ax.annotate('Fuel efficiency', xy=(9, 0.33), xytext=(12, 0.33), fontsize=40, color=colors[2], va='center')

ax.annotate('Before ML', xy=(0, 0), xytext=(0, -0.08), fontsize=40, color='black', ha='center')
ax.annotate('After ML', xy=(10, 0), xytext=(10.5, -0.08), fontsize=40, color='black', ha='center')

# Formatting
ax.axis('off')
plt.ylim(0, 1.2)
plt.xlim(-2, 14)
plt.tight_layout()
plt.show()
```

```python
ml_error = 5
current_predictions = 15
previous_predictions = 42

fig, ax = plt.subplots(figsize=(20, 10))

bar_height = 0.35
y_pos = [1, 0.5, 0]

ax.barh(y_pos[0], previous_predictions, color='lightgrey', height=bar_height)
ax.barh(y_pos[1], current_predictions, color='lightgrey', height=bar_height)
ax.barh(y_pos[2], ml_error, color='blue', height=bar_height)

ax.set_yticks(y_pos)
ax.set_yticklabels(['Previous predictions', 'Current predictions', 'Machine learning prediction'], fontsize=40)

ax.set_xlabel('Error of prediction method', fontsize=25, loc='right')
ax.set_xticks([])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.tight_layout()
plt.show()
```

```python

categories = [f'Fuel Consumption due\n to hard weather', 'Driver safety incidents', 'Weather-related delays']
reductions = [17, 42, 64]
remaining = [100 - r for r in reductions]
y = np.arange(len(categories))

fig, ax = plt.subplots(figsize=(20, 10))

ax.barh(y, [100] * len(categories), color=colors)
ax.barh(y, reductions, left=remaining, color='lightgrey')

for i in range(len(categories)):
    ax.text(100 - reductions[i] / 2, y[i], f'-{reductions[i]}%', 
            va='center', ha='center', color='black', fontsize=28)

ax.set_xticks([])
ax.set_yticklabels(categories, fontsize=25)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax.set_yticks(y)
ax.set_yticklabels(categories)
ax.set_xlim(0, 100)
ax.invert_yaxis()  

plt.tight_layout()
plt.show()
```
