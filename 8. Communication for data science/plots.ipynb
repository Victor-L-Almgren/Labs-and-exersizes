```python
import matplotlib.pyplot as plt

# Set global font to Arial
plt.rcParams['font.family'] = 'Calibri'
plt.rcParams['figure.facecolor'] = '"#FFFFFF"'


# Define the colors
colors = ["#69A2B0", "#9B6BAE", "#E8A87C"]

# Create the plot
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

# Plot the first line (Packages on time)
ax.plot([1, 9], [0.64, 0.78], color=colors[0], marker='o', linewidth=5, markersize=12)
ax.annotate('64%', xy=(1, 0.64), xytext=(-1.5, 0.64), fontsize=40, color=colors[0], va='center')
ax.annotate('78%', xy=(9, 0.78), xytext=(9.3, 0.78), fontsize=40, color=colors[0], va='center')
ax.annotate('Packages on time', xy=(9, 0.78), xytext=(12, 0.78), fontsize=40, color=colors[0], va='center')

# Plot the second line (Vehicle downtime) with flipped values
ax.plot([1, 9], [0.47, 0.56], color=colors[1], marker='o', linewidth=5, markersize=12)
ax.annotate('47%', xy=(1, 0.47), xytext=(-1.5, 0.47), fontsize=40, color=colors[1], va='center')
ax.annotate('56%', xy=(9, 0.56), xytext=(9.3, 0.56), fontsize=40, color=colors[1], va='center')
ax.annotate('Vehicle uptime', xy=(9, 0.56), xytext=(12, 0.56), fontsize=40, color=colors[1], va='center')

# Plot the third line (Fuel efficiency)
ax.plot([1, 9], [0.22, 0.33], color=colors[2], marker='o', linewidth=5, markersize=12)
ax.annotate('22%', xy=(1, 0.22), xytext=(-1.5, 0.22), fontsize=40, color=colors[2], va='center')
ax.annotate('33%', xy=(9, 0.33), xytext=(9.3, 0.33), fontsize=40, color=colors[2], va='center')
ax.annotate('Fuel efficiency', xy=(9, 0.33), xytext=(12, 0.33), fontsize=40, color=colors[2], va='center')

# Add Before and After labels
ax.annotate('Before ML', xy=(0, 0), xytext=(0, -0.08), fontsize=40, color='lightgrey', ha='center')
ax.annotate('After ML', xy=(10, 0), xytext=(10.5, -0.08), fontsize=40, color='lightgrey', ha='center')

# Add the title
ax.set_title('Performance Changes Before and After ML', fontsize=50, color='grey', ha='center', va='center')

# Formatting
ax.axis('off')
plt.ylim(0, 1.2)
plt.xlim(-2, 14)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

ml_error = 7
current_predictions = 15
previous_predictions = 42

fig, ax = plt.subplots(figsize=(20, 10))

bar_height = 0.35
y_pos = [1, 0.5, 0]

# Plot the bars
ax.barh(y_pos[0], previous_predictions, color='lightgrey', height=bar_height)
ax.barh(y_pos[1], current_predictions, color='lightgrey', height=bar_height)
ax.barh(y_pos[2], ml_error, color='blue', height=bar_height)

# Set y-axis labels
ax.set_yticks(y_pos)
ax.set_yticklabels(['Previous predictions', 'Current predictions', 'ML prediction'], fontsize=40)

# Set x-axis label
ax.set_xlabel('Relative prediction error', fontsize=25, loc='center')
ax.set_xticks([])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Annotate each bar
values = [previous_predictions, current_predictions, ml_error]
for y, value in zip(y_pos, values):
    ax.text(value + 1, y, f'{value}%', va='center', ha='left', fontsize=30)

plt.tight_layout()
plt.show()
```

```python
# Data
categories = [f'Fuel Consumption due\n to hard weather', 'Driver safety incidents', 'Weather-related delays']
reductions = [17, 42, 64]
y = np.arange(len(categories))

# Colors
orange = '#FFA500'  # orange hex code
grey = 'lightgrey'

bar_height = 0.4

fig, ax = plt.subplots(figsize=(20, 10))

# Plot the reduction (orange) bars
ax.barh(y - bar_height/2, reductions, color=orange, height=bar_height, label='After ML')

# Plot the remaining (grey) bars
ax.barh(y + bar_height/2, 100, color=grey, height=bar_height, label='Before ML')


# Annotate reductions
for i in range(len(categories)):
    ax.text(reductions[i] + 2, y[i] - bar_height/2, f'-{100-reductions[i]}%', 
            va='center', ha='left', color='black', fontsize=28)

# Set labels
ax.set_yticks(y)
ax.set_yticklabels(categories, fontsize=25)
ax.set_xticks([])
ax.set_xlim(0, 110)  # Extra space for text annotations
ax.invert_yaxis()  # Highest value on top


# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create time points
years = np.arange(2015, 2026)

# Generate base similar trend with small variations
np.random.seed(42)
base = np.linspace(100, 110, len(years)) + np.random.normal(0, 2, len(years))

# Create companies with slight differences
company_A = base + np.random.normal(0, 2, len(years))
company_B = base + np.random.normal(0, 2, len(years))
company_C = base + np.random.normal(0, 2, len(years))
company_D = base + np.random.normal(0, 2, len(years))

# Modify expenditures after 2018
for i, year in enumerate(years):
    if year > 2018:
        company_A[i] -= (year - 2018) * 10
        company_B[i] -= (year - 2018) * 12
        company_C[i] -= (year - 2018) * 11
        company_D[i] += (year - 2018) * 2  # Smaller increase now

# Plotting
fig, ax = plt.subplots(figsize=(10, 5))

# Add light grey rectangle from 2018 onward
ax.axvspan(2018, 2025, color='lightgrey', alpha=0.5)

# Plot companies A, B, C in grey
competitors = [company_A, company_B, company_C]
for idx, company in enumerate(competitors):
    ax.plot(years, company, color='grey', linewidth=2)
    # Annotate each competitor
    ax.text(years[-1] + 0.2, company[-1], f'Competitor {idx+1}', color='grey', fontsize=10, verticalalignment='center')

# Plot company D in blue
ax.plot(years, company_D, color='blue', linewidth=2)
# Annotate PostNord
ax.text(years[-1] + 0.2, company_D[-1], 'PostNord', color='blue', fontsize=12, verticalalignment='center')

# Final plot adjustments
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_ylabel('Normalized expenditures', fontsize=15)
ax.set_xticks(years)
ax.grid(False)
plt.tight_layout()
plt.show()
```

```python


# Create time points
years = np.arange(2015, 2026)

# Generate base similar trend with small variations
np.random.seed(42)
base = np.linspace(100, 110, len(years)) + np.random.normal(0, 2, len(years))

# Create companies with slight differences
company_A = base + np.random.normal(0, 2, len(years))
company_B = base + np.random.normal(0, 2, len(years))
company_C = base + np.random.normal(0, 2, len(years))
company_D = base + np.random.normal(0, 2, len(years))

# Different strong decline rates + random noise per year
np.random.seed(123)
decline_rates = {
    'A': 8 + np.random.uniform(-3, 3),
    'B': 13 + np.random.uniform(-3, 3),
    'C': 10 + np.random.uniform(-3, 3),
}

for i, year in enumerate(years):
    if year > 2018:
        company_A[i] -= (year - 2018) * decline_rates['A'] + np.random.normal(0, 3)
        company_B[i] -= (year - 2018) * decline_rates['B'] + np.random.normal(0, 3)
        company_C[i] -= (year - 2018) * decline_rates['C'] + np.random.normal(0, 3)
        company_D[i] += (year - 2018) * (2 + np.random.uniform(-1, 1))  # Small variation for PostNord

# Plotting
fig, ax = plt.subplots(figsize=(10, 5))

# Add light grey rectangle from 2018 onward
ax.axvspan(2018, 2025, color='lightgrey', alpha=0.5)

# Plot companies A, B, C in grey
competitors = [company_A, company_B, company_C]
for idx, company in enumerate(competitors):
    # Plot line
    ax.plot(years, company, color='grey', linewidth=2)
    # Add marker at end
    ax.plot(years[-1], company[-1], marker='o', color='grey')
    # Annotate with label and final value
    ax.text(years[-1] + 0.3, company[-1], f'Competitor {idx+1} ({company[-1]:.1f})', 
            color='grey', fontsize=15, verticalalignment='center')

# Plot company D in blue
ax.plot(years, company_D, color='blue', linewidth=2)
ax.plot(years[-1], company_D[-1], marker='o', color='blue')
ax.text(years[-1] + 0.3, company_D[-1], f'PostNord ({company_D[-1]:.1f})', 
        color='blue', fontsize=20, verticalalignment='center')

# Final plot adjustments
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_ylabel('Normalized expenditures', fontsize=15)
ax.set_xticks(years)
ax.grid(False)
plt.tight_layout()
plt.show()
```
