```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
from sklearn.decomposition import PCA
```

# Least Squares Fitting

```python
x = np.linspace(-5, 5, 201)
y = 1 + x + x * np.random.random(len(x))
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
# normalize using min-max normalization
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

```python
# assemble matrix A
A = np.concatenate([x, np.ones(len(x)).reshape(-1,1)],axis=1)
```

```python
# Direct least square regression
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),y)
print(alpha)
```

```python
# plot the results
plt.figure(figsize = (10,8))
plt.plot(x, y, 'b.')
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Calculate the MAE (mean absolute error)
y_pred = alpha[0]*x + alpha[1]
residual = y.ravel()-y_pred.ravel()
np.mean(np.abs(residual))
```

```python
# Plotting the residuals (error after fitting)
plt.plot(x,residual,'r.')
plt.xlabel('x values')
plt.ylabel('residuals')
```

# Exercise 1 

1- Do the following for four different degrees: 1, 2, 3, 4
- Use least square method to fit curve to the following dataset and plot the fitting curve. 
- Calculate MAE
- Plot the residuals

2- Compare the errors of fit for different degrees using a bar plot (x axis: degree, y axis: error)

```python
#Create the data
X_values = np.linspace(-10, 10, 201)
Y_values = 1 + X_values + X_values**2 + X_values**3 + 150*np.random.random(len(X_values))
X_values = X_values.reshape(-1,1)
Y_values = Y_values.reshape(-1,1)

#scaling the data
scaler = MinMaxScaler()
X_values = scaler.fit_transform(X_values)

#Plot the results
plt.figure(figsize = (10,5))
plt.plot(X_values, Y_values, 'b.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# degree 1 curve

x = X_values.copy()
y = Y_values.copy()
plt.figure(figsize=(15,5))

# creating the matrix A
A = np.concatenate([x, np.ones(len(x)).reshape(-1,1)],axis=1)

# Direct least square regression
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),y)

#calculate MAE
y_pred = alpha[0]*x + alpha[1]
residual = y.ravel()-y_pred.ravel()
mae_deg_1 = np.mean(np.abs(residual))

# plot of data
plt.subplot(1, 2, 1)
plt.plot(x, alpha[0]*x + alpha[1], 'red')
plt.plot(x, y, 'blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('degree 1 curve')

plt.subplot(1, 2, 2)
plt.plot(x, residual,'red')
plt.xlabel('x values')
plt.ylabel('residuals')
plt.title('plot of residuals')

plt.show()
```

```python
# degree 2 curve

x = X_values.copy()
y = Y_values.copy()
plt.figure(figsize=(15, 5))

# Creating the matrix A for quadratic regression
A = np.concatenate([x, x**2, np.ones(len(x)).reshape(-1, 1)], axis=1)

# Direct least square regression
alpha = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y)

# Calculate MAE
y_pred = alpha[0]*x**2 + alpha[1]*x + alpha[2]
residual = y.ravel() - y_pred.ravel()
mae_deg_2 = np.mean(np.abs(residual))

# Plot of data and regression line
plt.subplot(1, 2, 1)
plt.plot(x, alpha[0]*x**2 + alpha[1]*x + alpha[2], 'red')
plt.plot(x, y, 'blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('degree 2 curve')

# Plot of residuals
plt.subplot(1, 2, 2)
plt.plot(x, residual, 'red')
plt.xlabel('x values')
plt.ylabel('residuals')
plt.title('Plot of Residuals')

plt.show()
```

```python
# degree 3 curve

x = X_values.copy()
y = Y_values.copy()
plt.figure(figsize=(15, 5))

# Creating the matrix A for third-degree regression
A = np.concatenate([x, x**2, x**3, np.ones(len(x)).reshape(-1, 1)], axis=1)

# Direct least square regression
alpha = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y)

# Calculate MAE
y_pred = alpha[0]*x**3 + alpha[1]*x**2 + alpha[2]*x + alpha[3]
residual = y.ravel() - y_pred.ravel()
mae_deg_3 = np.mean(np.abs(residual))

# Plot of data and regression line
plt.subplot(1, 2, 1)
plt.plot(x, alpha[0]*x**3 + alpha[1]*x**2 + alpha[2]*x + alpha[3], 'red')
plt.plot(x, y, 'blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('degree 3 curve')

# Plot of residuals
plt.subplot(1, 2, 2)
plt.plot(x, residual, 'red')
plt.xlabel('x values')
plt.ylabel('residuals')
plt.title('Plot of Residuals')

plt.show()
```

```python
# degree 4 curve

x = X_values.copy()
y = Y_values.copy()
plt.figure(figsize=(15, 5))

# Creating the matrix A for third-degree regression
A = np.concatenate([x, x**2, x**3, x**4, np.ones(len(x)).reshape(-1, 1)], axis=1)

# Direct least square regression
alpha = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y)

# Calculate MAE
y_pred = alpha[0]*x**4 + alpha[1]*x**3 + alpha[2]*x**2 + alpha[3]*x + alpha[4]
residual = y.ravel() - y_pred.ravel()
mae_deg_4 = np.mean(np.abs(residual))

# Plot of data and regression line
plt.subplot(1, 2, 1)
plt.plot(x, alpha[0]*x**4 + alpha[1]*x**3 + alpha[2]*x**2 + alpha[3]*x + alpha[4], 'red')
plt.plot(x, y, 'blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('degree 3 curve')

# Plot of residuals
plt.subplot(1, 2, 2)
plt.plot(x, residual, 'red')
plt.xlabel('x values')
plt.ylabel('residuals')
plt.title('Plot of Residuals')

plt.show()
```

```python
# bar plot to compare errors

values = [mae_deg_1, mae_deg_2, mae_deg_3, mae_deg_4]
plt.bar(range(len(values)), values)
plt.ylabel('Value of MAE')
plt.title('Bar plot of MAEs')
plt.xticks(range(len(values)), ['Degree 1', 'Degree 2', 'Degree 3', 'Degree 4'])
plt.show()
```

# Excercise 2 (Multidimensional fit)

- Normalize the data using min-max normalization
- Use least square method to fit a second degree curve to the following dataset and plot the fitting curve
- Calculate MAE
- Which feature has the most contribution to predictions? (hint: you should understand based on the weights of features)

```python
from sklearn.datasets import load_diabetes
X = load_diabetes().data
y = load_diabetes().target
feat_names = load_diabetes().feature_names
df = pd.DataFrame(X,columns = feat_names)
```

```python
# normalize data
scaler = MinMaxScaler()

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y.reshape(-1, 1))
```

```python
# fit a second degree curve

# Fit a second degree curve
A = np.concatenate([X_scaled, X_scaled**2, np.ones((len(X_scaled), 1))], axis=1)

# Direct least square regression
alpha = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y_scaled)

# Predictions
y_pred = alpha[0]*X_scaled**2 + alpha[1]*X_scaled + alpha[2]

# Plotting
plt.scatter(X_scaled, y_scaled, color='blue', label='Actual Data')
plt.plot(X_scaled, y_pred, color='red', label='Second Degree Curve')
plt.xlabel('Normalized X-values')
plt.ylabel('Normalized Y-values')
plt.title('Second Degree Curve Fitting')
plt.legend()
plt.show()
```
