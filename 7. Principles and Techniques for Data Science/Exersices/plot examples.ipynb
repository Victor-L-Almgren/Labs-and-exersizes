```python
#Libraries
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

#Import data
label_encoder = LabelEncoder()
iris_df = sns.load_dataset('iris')
iris_df['target'] = label_encoder.fit_transform(iris_df['species'])

#Create a scatter plot
plt.figure(figsize=(10,8))
plt.scatter(iris_df['sepal_length'], iris_df['sepal_width'], 
            c=iris_df['target'], cmap='viridis', edgecolor='k')

#Adding axis lables and title
plt.xlabel('sepal length cm')
plt.ylabel('sepal wiwth cm')
plt.title('scatter plot of sepal length vs. width')

#Add a colorbar legend to 
plt.colorbar(label='iris spcies', ticks=[0, 1, 2], 
             orientation = 'vertical', boundaries=[-0.5, 0.5, 1.5, 2.5])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Import data
iris_df = sns.load_dataset('iris')
data = iris_df.to_numpy()
data = data[:, 0:4]  # Select only the first four columns (features)

# Create a histogram for each column (feature) in the dataset
plt.figure(figsize=(10, 8))
for i in range(data.shape[1]):
    plt.subplot(2, 2, i + 1)
    plt.hist(data[:, i], bins=20, color='orange', edgecolor='black')
    plt.xlabel(iris_df.columns[i])  # Use column names directly from the DataFrame
    plt.ylabel('Frequency')

plt.suptitle('Histograms of Iris Features')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Import data
iris_df = sns.load_dataset('iris')

# Create a KDE plot with different bandwidth adjustments
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

# Plot all features in each subplot with different bw_adjust values
sns.kdeplot(data=iris_df, x="sepal_length", fill=True, bw_adjust=0.5, ax=ax1, label="Sepal Length")
sns.kdeplot(data=iris_df, x="sepal_width", fill=True, bw_adjust=0.5, ax=ax1, label="Sepal Width")
sns.kdeplot(data=iris_df, x="petal_length", fill=True, bw_adjust=0.5, ax=ax1, label="Petal Length")
sns.kdeplot(data=iris_df, x="petal_width", fill=True, bw_adjust=0.5, ax=ax1, label="Petal Width")
ax1.set_title("bw_adjust = 0.5")
ax1.legend()

sns.kdeplot(data=iris_df, x="sepal_length", fill=True, bw_adjust=1, ax=ax2, label="Sepal Length")
sns.kdeplot(data=iris_df, x="sepal_width", fill=True, bw_adjust=1, ax=ax2, label="Sepal Width")
sns.kdeplot(data=iris_df, x="petal_length", fill=True, bw_adjust=1, ax=ax2, label="Petal Length")
sns.kdeplot(data=iris_df, x="petal_width", fill=True, bw_adjust=1, ax=ax2, label="Petal Width")
ax2.set_title("bw_adjust = 1")
ax2.legend()

sns.kdeplot(data=iris_df, x="sepal_length", fill=True, bw_adjust=1.5, ax=ax3, label="Sepal Length")
sns.kdeplot(data=iris_df, x="sepal_width", fill=True, bw_adjust=1.5, ax=ax3, label="Sepal Width")
sns.kdeplot(data=iris_df, x="petal_length", fill=True, bw_adjust=1.5, ax=ax3, label="Petal Length")
sns.kdeplot(data=iris_df, x="petal_width", fill=True, bw_adjust=1.5, ax=ax3, label="Petal Width")
ax3.set_title("bw_adjust = 1.5")
ax3.legend()

# Add an overall title
plt.suptitle("Effect of bw_adjust on KDE for Iris Dataset Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from mpl_toolkits.mplot3d import Axes3D  # For 3D projection

# Load and prepare the data
label_encoder = LabelEncoder()
iris_df = sns.load_dataset("iris")
y = label_encoder.fit_transform(iris_df['species'])  # Encode species labels
iris_df = iris_df.drop('species', axis=1)  # Remove the species column
feature_names = list(iris_df.columns.values)  # Get feature names
data = iris_df.to_numpy()  # Convert DataFrame to NumPy array

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

colors = ['r', 'g', 'b']
markers = ['o', 's', '^']

# Plot each species with different color and marker
for i in range(3):  # Since we have 3 species in the iris dataset
    ax.scatter(data[y == i][:, 0], data[y == i][:, 1], data[y == i][:, 2],
               c=colors[i], marker=markers[i], label=f'Species {i+1}')

# Set axis labels using feature names
ax.set_xlabel(feature_names[0])
ax.set_ylabel(feature_names[1])
ax.set_zlabel(feature_names[2])

# Add legend and title
ax.legend()
plt.title("3D Visualization of Iris Dataset")
plt.show()
```

```python
import seaborn as sns
import plotly.express as px
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
iris_df = sns.load_dataset("iris")
iris_df['target'] = label_encoder.fit_transform(iris_df['species'])

# Create an interactive 3D scatter plot using Plotly
fig = px.scatter_3d(iris_df, x='petal_width', y='sepal_width', z='sepal_length',
                    color='target', symbol='target', opacity=0.7,
                    title='Interactive 3D Scatter Plot of Iris Dataset',
                    labels={'target': 'Iris Species'})

fig.update_traces(marker=dict(size=5))
fig.show()
```
