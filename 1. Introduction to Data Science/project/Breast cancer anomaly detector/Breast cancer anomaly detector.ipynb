```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import math
from sklearn.neighbors import KernelDensity
from sklearn.neighbors import LocalOutlierFactor
```

```python
# A) Data exloration

print(f"*** Data Exploration *** \n")
data = pd.read_csv("patients.csv")
print(f"Shape of patients.csv is: {data.shape} \n")
print(f"The mean of the different features are as follows: \n")
print(data.mean())
print()
print(f"The standard devation of the different features are as follows:  \n")
print(data.std())
print()

print(f"Histogram of a given feature in patients.csv: \n")
feature = data["worst area"]
plt.hist(feature, color="green")
plt.ylabel("nr of datapoints")
plt.xlabel("value of datapoint")
plt.show()

print()
print(f"The initial rows of Patients.csv looks like this: \n")
data.head(10)
```

```python
# B) Data preprocessing and visualization
print(f"*** Data preprocessing and visualization *** \n")

scaler = StandardScaler()
numpy_data = data.to_numpy()
scaler.fit(numpy_data)
data_scaled = scaler.transform(numpy_data)

pca2 = PCA(n_components=2)
pca2.fit(data_scaled)
data_pca_transformed = pca2.transform(data_scaled)
plt.scatter(data_pca_transformed[:, 0], data_pca_transformed[:, 1], color ="green", marker ="x",label="datapoints")
plt.title("Plot of scaled and PCA-transformed data from patients.csv")
plt.legend()
plt.show()

kde = KernelDensity()
kde.fit(data_pca_transformed)
log_density = kde.score_samples(data_pca_transformed)
density = np.exp(log_density)
ids = np.argsort(density)
ids = ids[:30]
plt.scatter(data_pca_transformed[:, 0], data_pca_transformed[:, 1], color ="green", marker ="x",label="normal datapoints")
plt.scatter(data_pca_transformed[ids, 0], data_pca_transformed[ids, 1], color="red", marker ="x", label="potential anomalies identified with KDE")
plt.legend()
plt.show()

lof = LocalOutlierFactor(n_neighbors=10)
lof.fit(data_pca_transformed)
scores = lof.negative_outlier_factor_
ids = np.argsort(scores)
ids = ids[:30]
plt.scatter(data_pca_transformed[:, 0], data_pca_transformed[:, 1], color ="green", marker ="x",label="normal datapoints")
plt.scatter(data_pca_transformed[ids, 0], data_pca_transformed[ids, 1], color="red", marker ="x", label="potential anomalies identified with LOF")
plt.legend()
plt.show()
```

```python
# C) Calculating Euclidean Distance
print(f"*** Calculating Euclidean Distance *** \n")

def euclidean_distance(p1, p2):
    p1_array = np.array(p1)
    p2_array = np.array(p2)
    distance = (sum((p1_array - p2_array)**2))**0.5
    return distance

p1 = data_pca_transformed[0, 0], data_pca_transformed[0, 1]
p2 = data_pca_transformed[1, 0],data_pca_transformed[1, 1]
test = euclidean_distance(p1, p2)
print(f"Euclidean distance between point {p1} and {p1} is {test}")
```

```python
# D) Anomaly Score Calculation
print(f"*** Anomaly Score Calculation *** \n")

def calculate_anomaly_score(data, query_point, k):
    all_distances = []
    for datapoint in range(len(data)):
        euc_dist = euclidean_distance(data[datapoint], query_point)
        all_distances.append(euc_dist)
    all_distances.sort()
    all_distances = np.array(all_distances)
    anomaly_score = (all_distances[1:k+1]).sum()/k
    return anomaly_score

query_point = data_pca_transformed[0]
k = 5
test = calculate_anomaly_score(data_pca_transformed, query_point, k)
print(f"Average distance from query point {query_point} to closest {k} neighbors is {test}")
```

```python
# E) Identifying Anomalies
print(f"*** Identifying Anomalies *** \n")

def get_all_anomaly_scores(data, k):
    all_anomaly_scores = []
    for datapoint in range(len(data)):
        anomaly_score = calculate_anomaly_score(data, data[datapoint], k)
        all_anomaly_scores.append(anomaly_score)
    return all_anomaly_scores

def get_top_anomaly_scores(all_anomaly_scores, n_of_scores):
    sorted_data = sorted(all_anomaly_scores)
    top_anomaly_scores = []
    while n_of_scores:
        top_anomaly_scores.append(sorted_data[-n_of_scores])
        n_of_scores -= 1
    return top_anomaly_scores

def get_anomaly_patients(all_anomaly_scores, top_anomaly_scores, data):
    anomaly_patients_index = []
    anomaly_patients = []
    for index in top_anomaly_scores:
        anomaly_patients_index.append(all_anomaly_scores.index(index))
    for index in anomaly_patients_index:
        anomaly_patients.append(data[index])
    anomaly_patients = np.array(anomaly_patients)
    return anomaly_patients          

all_anomaly_scores = get_all_anomaly_scores(data_pca_transformed, 10)
print(f"All anomaly scores are: \n {all_anomaly_scores} \n")

top_anomaly_scores = get_top_anomaly_scores(all_anomaly_scores, 30)
print(f"The top 28 anomaly scores are: \n {top_anomaly_scores} \n")

anomaly_patients = get_anomaly_patients(all_anomaly_scores, top_anomaly_scores, data_pca_transformed)
print(f"The top 28 anomaly patients are: \n {anomaly_patients}")
```

```python
# F) Visualization of Anomalies
print(f"*** Visualization of Anomalies *** \n")

plt.scatter(data_pca_transformed[:, 0], data_pca_transformed[:, 1], color ="green", marker ="x", label="datapoints")
plt.scatter(anomaly_patients[:, 0], anomaly_patients[:, 1], color="red", marker="x", label="anomaly patients")
plt.title("Plot of scaled and PCA-transformed data from patients.csv")
plt.legend()
plt.show()
```
