# Introduction to Regression
In this Lab, you will use linear regression with one feature to predict profits for a food truck. Suppose you are the CEO of a restaurant franchise and are considering different cities for opening a new outlet. Your company already has trucks in various cities and you have data for profits and populations from the cities. You would like to use this data to help you select which city to expand to next.

We have a file `datasets/PopulationProfit.csv` that contains the dataset for our linear regression problem. Complete the code below to load the dataset into a pandas DataFrame called `df`, then display the first 5 rows from it.

```python
""" TODO: 
load the dataset from "datasets/PopulationProfit.csv into 
a pandas DataFrame then display the first 5 rows from it.
"""

import pandas as pd

df = pd.read_csv("datasets/PopulationProfit.csv")
print(df[0:5])
```

The first column (called `"population"`) is the population of a city and the second column (called `"profit"`) is the profit of a food truck in that city. 

- To select one column from `df` and get its values into a simple one-domensional array, you can use `df[column_name].values`. For example, `df[ "profit" ].values` gives a simple array containing the profit values.
- To select one or several columns from `df` and get their values into a matrix (two-dimensional array), you can use `df[list_of_columns]` (for example, `df[ ["population"] ].values` gives a matrix with one column corresponding to the population values.

Complete the code below to extract the data from the DataFrame `df` into the variables `X` and `y`. The variable `X` is the input data and should be a matrix (two-dimensional array) with one column (one feature) corresponding to the population. The variable `y` (one-dimensional array) contains the output values corresponding to the profit. 

```python
#""" TODO: 
#Extract the data from df into the variables X and y as described above.
#"""

X = df[["population"]].values
y = df["profit"].values


#""" TODO:
#Print the shape of the variables X and y. You should see 
#that X have shape (97, 1), and y have shape (97,).
#"""

print(X.shape)
print(y.shape)
```

Before applying machine learning, it is often useful to understand the data by visualizing it. Here, you can use a scatter plot to visualize the data, since it has only two properties to plot (population and profit). Complete the *TODO*s in the code below to produce this scatter plot. Make sure that the result of your plot looks like the following figure, with the same red ${\color{red}+}$ markers and axis labels.

```python
""" TODO:
Complete the lines below to do a scatter plot showing 
the population vs. the profit in various cities.
"""

import matplotlib.pylab as plt

plt.scatter(X, y, color="red", marker="+")
plt.ylabel("Profit in $10.000s")
plt.xlabel("Population in 10.000s")
plt.show()
```

Next, you will use the *scikit-learn* machine learning library to perform a linear regression on the previous data. Complete the code below to train a linear regression model `LinearRegression`.

```python
""" TODO:
Complete the code to train the LinearRegression model 
using the training dataset (input data X and output y).
"""

from sklearn.linear_model import LinearRegression

reg_model = LinearRegression()  # Creating the LinearRegression model (untrained)
# TODO: train the reg_model using the training dataset
reg_model.fit(X, y)
```

Finally, use your trained regression model to make predictions on profits in areas of 35,000 and 70,000 people. In other words, you are asked to predict the profit for a new dataset `X_test` with two rows: `[3.5]` (area with a population of 35,000 people) and `[7]` (area with a population of 70,000 people).

```python
""" TODO:
Complete the code to predict the profit in areas of 35,000 and 70,000 people
"""
import numpy as np

# Our test data is:
X_test = np.array([
    [3.5], # area with a population of 35,000 people
    [7]    # area with a population of 70,000 people
])

# TODO: make the predictions

y_pred = reg_model.predict(X_test)
for earnings in y_pred:
        print("Approx: ", int(earnings * 10000), "USD")
```

In addition to the training data you plotted previously in red, you can now also plot the test data (with their predicted profit) in a different color (e.g. green). Here is an example:

```python
""" TODO:
In addition to the previous training data
Plot the test data and the predicted profit 
"""

import matplotlib.pylab as plt

plt.scatter(X, y, color="red", marker="+", label="Training data")
plt.plot(X_test, y_pred, color="green", label="Test data")
plt.legend()
plt.ylabel("Profit in $10.000s")

plt.xlabel("Population in 10.000s")
plt.show()
```
