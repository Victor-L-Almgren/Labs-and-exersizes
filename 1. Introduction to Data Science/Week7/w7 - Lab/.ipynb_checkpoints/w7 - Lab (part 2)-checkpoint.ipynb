## 2. Extracting Features from Emails
After preprocessing the email, the next step is to implement the feature-extraction that represents the preprocessed email as a feature-vector (containing numbers instead of text).

Before implementing the feature-extraction, we first need a vocabulary list of all possible words. The complete vocabulary list is in the file `datasets/vocab.txt` (one word per line), it contains 1899 words. In the Python code below, the vocabulary list (of 1899 of words) is already loaded for you in the variable `vocabulary`.

Based on this vocabulary list, we can now represent a preprocessed email as a feature-vector of $1899$ features. The $j^{th}$ feature indicates whether or not the $j^{th}$ vocabulary word is present in the email or not (1 if the word is present, and 0 if not).

To better understand this, let's have a simple example. To simplify this example, suppose that out vocabulary list consists of 8 words as follows: `vocabulary = ["about", "anyon", "bitcoin", "bank", "each", "site", "where", "web"]`, and that the content of our preprocessed email is just: `"anyon know how much it cost to host web portal"`. Then, this email can be represented with the following feature-vector (of 8 values): `x = [0, 1, 0, 0, 0, 1, 0, 1]`. As you can see, the feature `x[j]` for an email corresponds to whether the word `vocabulary[j]` occurs in the email.

You should now complete the following Python code to generate a feature-vector (of 1899 values) for an email, based on the loaded vocabulary list.

**Hints:** 
- **Splitting a string:** In Python, when you have some text (as a string), for example `text = "Hi, no, it is not sunny, is it ?"`, you can convert it to a list of words (list of strings) by splitting it based on the space character `" "` as follows: `list_of_words = text.split(" ")`. In this case, `list_of_words` would be `["Hi,", "no,", "it", "is", "not", "sunny,", "is", "it", "?"]`. Another example: `text.split(", ")` would result in `["Hi", "no", "it is not sunny", "is it ?"]`.

- **Converting a list to a set:** You can convert a list to a set of unique items. For example, if we have the list `list_of_words = ['Hi,', 'no,', 'it', 'is', 'not', 'sunny,', 'is', 'it', '?']`, we can get a set of unique items using `set_of_words = set(list_of_words)`. In this case, `set_of_words` would be `{'it', 'Hi,', 'sunny,', 'no,', '?', 'not', 'is'}`. Notice that the items `"it"` and `"is"` are unique in the set (not repeated). Also, notice that unlike a list, the items in a set are not ordered (they can appear in any order).

- **Check if an item exists in a set of items:** You can check in a fast way if a given string `word` exists in a set `set_of_words` by testing if `word in set_of_words`. In the previous example, `"sunny" in set_of_words` would be `True`, but `"coffee" in set_of_words` would be `False`. 

```python
import numpy as np

# This loads a vocabulary list of 1899 words
vocabulary = [ word.strip() for word in open("datasets/vocab.txt") ]
vocabulary = np.array(vocabulary) # Convert it to an array


""" TODO:
Write the definition of the email_to_vector function which takes as arguments: the vocabulary 
array of words, and the raw email text (original_email). The function should return an array of 
the same size as the vocabulary array, where the j-th element of this array indicates whether or not 
the j-th vocabulary word is present in the preprocessed version of original_email. 
"""
def email_to_vector(vocabulary, original_email):
    # TODO: process the original_email using the processEmail function
    preprocessed_email = processEmail(original_email)
    
    # TODO: split the preprocessed_email text to convert it to a list of words
    email_words = preprocessed_email.split(" ")
    
    # TODO: convert the list email_words to a set email_words_set
    email_words_set = set(email_words)
    
    """
    Next, you need to create a list of the same length as the vocabulary and 
    populate it with values corresponding to 1 if the vocabulary word exists 
    in the email_words_set and 0 otherwise. Follows the bellow steps to do this.
    """
    feature_vector = [] # We can start with an empty list
    # TODO: for each word in vocabulary, append either 1 or 0 to feature_vector ...
    # ... depending on whether the word exists or not in email_words_set
    for word in vocabulary:
        if word in email_words_set:
            feature_vector.append(1)
        else:
            feature_vector.append(0)
    
    # We convert the feature_vector to an array and return it
    return np.array(feature_vector)


""" TODO:
To test your email_to_vector(..) function, call it using the sample_email text defined previously.
Count how many entries in the resulting feature-vector are set to 1? It should be around 43 among 1899.
"""
test = email_to_vector(vocabulary, sample_email)
n = 0
for position in test:
    if position == 1:
        n+=1
print(n)
```

## 3. Training SVM for Spam Classification
In this section, we will load the training dataset we prepared for you; it will be used to train a SVM classifier and test its performance on new emails.

The data file `spamTrain.csv` contains 4000 training examples (as rows) corresponding to various emails, while the `spamTest.csv` file contains 1000 test examples. Each original email was processed for you using the `email_to_vector(..)` function and converted into a feature vector (of 1899 features). The last column in these `.csv` files corresponds to the output which indicates if an email is a *spam* (label 1) or *non-spam* (label 0).

```python
import pandas as pd

"""TODO:
Read and and try to understand the code below, then run it.
"""

# Loading the training dataset into X (inputs) and y (outputs)
df_train = pd.read_csv("datasets/spamTrain.csv")
X = df_train.iloc[:, :-1].values # Take all columns except the last one
y = df_train.iloc[:, -1].values  # Take the last column (corresponding to the class-label)
print("Training dataset ==> X.shape:", X.shape, "y.shape:", y.shape)

# Loading the testing dataset into Xtest (inputs) and ytest (outputs)
df_test = pd.read_csv("datasets/spamTest.csv")
Xtest = df_test.iloc[:, :-1].values # Take all columns except the last one
ytest = df_test.iloc[:, -1].values  # Take the last column (corresponding to the class-label)
print("Testing dataset ==> Xtest.shape:", Xtest.shape, "ytest.shape:", ytest.shape)
```

Now let's try to visualize this data. 
As you probably know, we CAN NOT directly visualize this dataset since we have 1899 features (not just 2 features). However, as you also know, we can use a technique called PCA (Principal Componants Analysis) to transform and reduce this high-dimensional data to a low-dimensional data (e.g. characterized by 2 features or components) and be able to visualize it. The visualization corresponds to the following figure. Note that this visualization does not capture all the information contained in the original 1899 features, but it captures most of it.

<img src="imgs/spamDatasetPCALab5B.png" width="500px" />

```python
"""TODO:
Read and and try to understand the code below, then run it.
"""

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Transforming the original input data using PCA
X_pca = PCA(n_components = 2).fit_transform(X)
print(f"shape of original data: {X.shape}, shape of transformed data: {X_pca.shape}")

# Separate non-spam data and the spam data to visualize them with different colors
X0_pca = X_pca[y==0] # Rows of X_pca where the class-label is 0 (non-spam emails)
X1_pca = X_pca[y==1] # Rows of X_pca where the class-label is 1 (spam emails)

# Visualize the data
plt.scatter(X0_pca[:, 0], X0_pca[:, 1], marker=".", color="green", label="Non-spam (label 0)")
plt.scatter(X1_pca[:, 0], X1_pca[:, 1], marker="1", color="red", label="Spam (label 1)")

plt.xlabel("First principal componant of PCA")
plt.ylabel("Second principal componant of PCA")
plt.title("Training dataset visualization using PCA")

plt.legend()
plt.show()
```

After loading the dataset, the following Python code will proceed to train a linear SVM with the hyper-parameter $C=0.03$ to classify between spam and non-spam emails.

Complete the code to compute the accuracy on the training set and the accuracy on the test set. Once the training completes, you should see that the classifier gets a training accuracy of about 99.4% and a test accuracy of about 99.0%.

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Training
model = SVC(C=0.03, kernel="linear")
model.fit(X, y)

# TODO: predict the output corresponding to the training data X
y_pred = model.predict(X)

# TODO: predict the output corresponding to the test data Xtest
ytest_pred = model.predict(Xtest)

# TODO: Using the accuracy_score function imported above, compute the following accuracies:
acc_train = accuracy_score(y, y_pred)
acc_test = accuracy_score(ytest, ytest_pred)
print(f"Training Accuracy = {acc_train}, Test Accuracy = {acc_test}")
```

## 4. Top Predictors for Spam
You have almost finished with this project. This section is mostly for information, and you don't need to implement anything here.

To better understand how the spam classifier works, we can inspect the parameters learned by the linear SVM to see which words (i.e. features) the classifier thinks are the most important for predicting spams (class 1). To do so, we need to find the parameters having the largest values and display the vocabulary words corresponding to them. You can read the following code. It finds the 10 words associated with the highest parameter values in the classifier.

In our case, we found out that if an email contains words such as 'click', 'remov', 'basenumb', 'our', 'guarante', 'visit', 'pleas', 'you', 'dollar', and 'here', it is likely to be classified as spam. In other words, the classifier "thinks" that these words are the most likely indicators of spam.

```python
# Get the vector of parameters learned by the linear classifier
params = model.coef_.flatten()

# Get the indices that would sort the params array (from smallest to highest)
ids = np.argsort(params)

# Pick the 10 indices corresponding to the largest parameters (the 10 last ones)
top_ids = ids[-10:]

# Pick the vocabulary words corresponding to these indices
top_words = vocabulary[top_ids]
top_words
```
