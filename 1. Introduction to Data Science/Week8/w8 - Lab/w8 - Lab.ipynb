# K-means Clustering

The KMeans clustering method is used to automatically cluster similar data-points together. In this assignment, you will apply KMeans to compress an image (by clustering its pixel colors).

You will first start on a simple example dataset that will help you gain an intuition of how KMeans works. After that, you will use KMeans for image compression by reducing the number of colors that occur in an image to only those that are most common in that image.

## 1. KMeans on an example dataset
Let's start by using KMeans on a simple artificial dataset to help you understand how it works. The dataset is stored in the file `datasets/simpleDataset.csv` and have two features (columns). Run the following code to load the dataset and complete the code to plot it. Note that the dataset is unlabeled, so we only have an input matrix `X` (without labels `y`).

```python
import pandas as pd
import matplotlib.pylab as plt

df = pd.read_csv("datasets/simpleDataset.csv")
X = df.values
print("X.shape:", X.shape)

""" TODO:
Produce a scatter plot of the data (use "." as the marker and "grey" as the color. 
"""
plt.scatter(X[:, 0], X[:, 1], color="grey", marker=".")
plt.show()
```

Next, use KMeans (from scikit-learn) to find three clusters in the dataset `X` by following the *TODOs* in the code below.

Once the clustering is done and the cluster-labels are predicted, your task is to produce a plot of the dataset with colors corresponding to the clusters that KMeans found. Your plot should be similar to the following figure.
<img src="imgs/clustResultLab7A.png" width="500px" />
**Hint:** In order, for example, to select the data-points that are members of cluster 0, you can use `X[idx == 0]`, where `idx` corresponds to the array of predicted labels (each label corresponds to a cluster index: 0, 1, or 2).


```python
from sklearn.cluster import KMeans


""" TODO:
Train KMeans on the dataset X. The number of clusters to 
find should be set to 3.
"""
model = KMeans(n_clusters = 3)
idx = model.fit_predict(X)

""" TODO:
Predict the labels on the dataset X. These labels would 
correspond to the indices of clusters (0, 1, 2)
"""
idx0 = X[idx == 0]
idx1 = X[idx == 1]
idx2 = X[idx == 2]


""" TODO:
Produce the required plot
"""
plt.scatter(idx0[:, 0], idx0[:, 1], color="purple", marker="o", label="Cluster 0")
plt.scatter(idx1[:, 0], idx1[:, 1], color="green", marker="<", label="Cluster 1")
plt.scatter(idx2[:, 0], idx2[:, 1], color="blue", marker="*", label="Cluster 2")
plt.title("KMeans")
plt.legend()
plt.show()
```

## 2. Image compression with KMeans
<img src="datasets/bird_small.png" />

In this section, you will apply KMeans to image compression. In a straightforward 24-bit color representation of an image, each pixel is represented as three 8-bit unsigned integers (ranging from 0 to 255) that specify the red, green and blue intensity values. This encoding is often referred to as the RGB encoding. Our image contains thousands of colors, and in this part of the project, you will reduce the number of colors to only 16 colors. By making this reduction, it is possible to represent (compress) the photo in an efficient way.

In this section, you will use KMeans to select the 16 colors that will be used to represent the compressed image. Concretely, you will treat every pixel (color) in the original image as a data-point and use KMeans to find the 16 colors that best group (cluster) the pixels in the 3-dimensional RGB space. Once you have computed the cluster centroids on the image, you will then use the 16 colors to replace the pixels in the original image.

### 2.1. K-means on pixels
In Python, images can be read as follows `A = plt.imread('image.png')`. For RGB images, this function creates a three-dimensional array $A$ of shape `(height, width, 3)` where $height \times width$ is the number of pixels in the image, and each element `A[i][j]` (corresponding to the pixel at position (i, j) of the image) is a vector of three values containing the RGB (*Red*, *Green*, *Blue*) intensities.

We consider each pixel as a data-point (with three features). Therefore, the number of data-points we have is the number of pixels in the image (i.e. $n = height \times width$). For example, for an RGB image of $n = 128 \times 128 = 16384$ pixels, our dataset is `X`would contain 16384 rows and 3 columns.

The following Python code first loads the image for you, and then reshapes it to create a dataset `X`. Read and understand the code, then run it.

```python
import matplotlib.pylab as plt

# Loading the image bird_small.png into a three dimensional array A (with shape (128, 128, 3))
A = plt.imread('datasets/bird_small.png')
print("A.shape:", A.shape)

# Reshape A into a matrix X of (128*128) rows and 3 columns
X = A.reshape(A.shape[0] * A.shape[1], A.shape[2])
print("X.shape:", X.shape)

# Just showing the original image (loaded in A)
plt.imshow(A)
plt.show()
```

Now you have a dataset `X` where each row correspond to a pixel color (with 3 values). Your task in the code below is to use KMeans to cluster these colors into 16 clusters. Follow the instructions in the *TODOs*. <br>

Once you finish implementing the *TODOs*, you can run the cell, and see the results. The new image is represented using the centroids colors instead of the original color. The image that you will get is shown in the following figure. 
<img src="imgs/KmeansCompressedBird.png" width="350px" />

Finally, you can try to re-run the code below with a smaller number of clusters (e.g. 10, 5, 2 ...) instead of 16, and see the results again.

```python
import numpy as np

print(X.shape)
# =======================================
# TODO: Train KMeans on X to cluster the pixels into 16 clusters
km_model = KMeans(n_clusters=16)
km_model.fit(X)

# TODO: Predict the labels (cluster indices) for all the pixels in X
idx = km_model.predict(X)

# TODO: Uncomment the next line to get the cluster centers (centroids)
centroids = km_model.cluster_centers_

# =======================================
# TODO: Create a new matrix XX (having the shape as X) where each data-point (row) is replaced by its cluster centroid 
# (i.e. the centroid of the cluster that the data-point belongs to)
lst = []
for i in range(len(idx)):
    lst.append(centroids[idx[i]])
XX = np.array(lst)

# Reshaping XX back into an image (3-dimensional array shape (128,128,3))
AA = XX.reshape(128, 128, 3)

# Ploting the original image A and the new image AA
plt.imshow(A)
plt.axis("off")
plt.title("Original image")
plt.show()

plt.imshow(AA)
plt.axis("off")
plt.title("Compressed image")
plt.show()
```
