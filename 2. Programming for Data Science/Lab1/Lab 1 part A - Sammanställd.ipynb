```python
#Imported libraries
from faker import Faker
import time
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Task 1 - Bubble Sort
def bubble_sort(lst, column_name):  
    is_sorted = False   # Flag to track whether the list is sorted
    while not is_sorted:    # Continue looping until the list is sorted
        is_sorted = True    # Assume the list is sorted at the beginning of each pass
        for x in range(len(lst)-1):     # Iterate through each pair of adjacent elements in the list
            if lst[x][column_name] > lst[x+1][column_name]:      # Compare the values of the specified column for adjacent elements
                lst[x], lst[x+1] = lst[x+1], lst[x]      # If they are out of order, swap them
                is_sorted = False   # Since a swap was made, the list is not fully sorted
    return(lst)  # Return the sorted list

fake = Faker()
lst = []
for x in range(100):
    lst.append(fake.profile())
test = bubble_sort(lst, "ssn")
pd.DataFrame(test)
```

```python
#Task 2 - Selection Sort
def selection_sort(lst, column_name):   
    for x in range(len(lst)):   # Iterate through each element in the list
        min_index = x   # Assume the current index is the minimum
        for y in range(x+1, len(lst)):    # Iterate through the remaining elements in the list
            if lst[y][column_name] < lst[min_index][column_name]:    # Compare values based on the specified column
                min_index = y   # Update the index of the minimum value
        lst[x], lst[min_index] = lst[min_index], lst[x]     # Swap the current element with the minimum element found
    return(lst)  # Return the sorted list

fake = Faker()
lst = []
for x in range(100):
    lst.append(fake.profile())
test = selection_sort(lst, "ssn")
pd.DataFrame(test)
```

```python
#Task 3 - Merge sort
def merge_sorting(lst, column_name):
    if len(lst) > 1:
        mid = len(lst) // 2
        left_lst = lst[:mid]
        right_lst = lst[mid:]

        merge_sorting(left_lst, column_name)
        merge_sorting(right_lst, column_name)

        i = j = k = 0

        while i < len(left_lst) and j < len(right_lst):
            if left_lst[i][column_name] < right_lst[j][column_name]:
                lst[k] = left_lst[i]
                i += 1
            else:
                lst[k] = right_lst[j]
                j += 1
            k += 1

        while i < len(left_lst):    # Copy the remaining elements of left_lst, if any
            lst[k] = left_lst[i]
            i += 1
            k += 1
            
        while j < len(right_lst):   # Copy the remaining elements of right_lst, if any
            lst[k] = right_lst[j]
            j += 1
            k += 1
            
    return(lst)

fake = Faker()
lst = []
for x in range(100):
    lst.append(fake.profile())
test = merge_sorting(lst, "ssn")
pd.DataFrame(test)
```

```python
#Task 4 - Comparison
def get_avg_times(size_of_list, n_itterations, column_name):
    # Lists to store time measurements for each sorting algorithm
    time_list_bubble = []
    time_list_select = []
    time_list_merge = []
    time_list_sort = []
    
    # Create a Faker instance for generating fake profiles
    fake = Faker()
    
    # Run the sorting algorithms for the specified number of iterations
    for x in range(n_itterations):
        # Generate a list of fake profiles
        lst = []
        for x in range(size_of_list):
            lst.append(fake.profile())
            
        # Measure time for bubble sort
        t0 = time.time()
        test = bubble_sort(lst, column_name)
        t_bubble = time.time() - t0
        time_list_bubble.append(t_bubble)

        # Measure time for selection sort
        t0 = time.time()
        test = selection_sort(lst, column_name)
        t_select = time.time() - t0
        time_list_select.append(t_select)

        # Measure time for merge sort
        t0 = time.time()
        test = merge_sorting(lst, column_name)
        t_merge = time.time() - t0
        time_list_merge.append(t_merge)

        # Measure time for built-in sort
        t0 = time.time()
        sorted_list = sorted(lst, key=lambda x: x[column_name])
        t_builtin = time.time() - t0
        time_list_sort.append(t_builtin)

    # Calculate average times for each sorting algorithm
    avg_bubble = sum(time_list_bubble) / n_itterations
    print(f"Bubble sort time was: {avg_bubble} seconds")
        
    avg_select = sum(time_list_select) / n_itterations
    print(f"Selection sort time was: {avg_select} seconds")

    avg_merge = sum(time_list_merge) / n_itterations
    print(f"Merge sort time was: {avg_merge} seconds")

    avg_sort = sum(time_list_sort) / n_itterations
    print(f"Built-in sort time was: {avg_sort} seconds")

    # Return a list containing the average times for each sorting algorithm
    return [avg_bubble, avg_select, avg_merge, avg_sort]

test = get_avg_times(1000, 10, "ssn")
positions = range(len(test))
labels = ["Bubble sort", "Selection sort", "Merge sort", "Built in sort"]
plt.bar(positions, test)
plt.ylabel("Time in seconds")
plt.xlabel("Sorting algorithms")
plt.xticks(positions, labels)
plt.show()
```
