```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
#%matplotlib notebook
```

# Different types of edges

```python
x_step = np.linspace(0,10,100)
y_step = [0 if  i <5 else 1 for i in x_step]
plt.plot(x_step, y_step)
plt.title('Step function')
plt.show()
```

# Exercise 1

1. Take the discrete first-order derivative of the aforementioned step function and plot the result
2. Take the discrete second-order derivative of the aforementioned step function and plot the result
3. Add unifrom noise with magnitude 0.05,0.1, 0.2 to the function and perform part 1 and 2 for the resulting function

```python
#Anwser exersice 1.1
delta_x_step = x_step[1]-x_step[0]
y_step_d1 = np.diff(y_step) / delta_x_step

plt.figure(figsize=(12, 8))
plt.axhline(y=0, c='black')
plt.axvline(x=0, c='black')
plt.grid()

plt.plot(x_step[0:-1], y_step_d1, label = 'first order derivative')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('First order derivative of step function')
plt.legend()
plt.show()
```

```python
#Anwser exersice 1.2
y_step_d2 = np.diff(y_step_d1) / delta_x_step

plt.figure(figsize=(12, 8))
plt.axhline(y=0, c='black')
plt.axvline(x=0, c='black')
plt.grid()

plt.plot(x_step[0:-2], y_step_d2, label = 'second order derivative', c='orange')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Second order derivative of step function')
plt.legend()
plt.show()
```

```python
#Awnser exersice 1.3

#Adding noise
noise_1 = np.random.uniform(-0.025, 0.025, x_step.shape)
noise_2 = np.random.uniform(-0.05, 0.05, x_step.shape)
noise_3 = np.random.uniform(-0.1, 0.1, x_step.shape)

y_step_n1 = []
y_step_n2 = []
y_step_n3 = []

ind = 0
for i in x_step:
    if i < 5:
        y_step_n1.append(noise_1[ind])
        y_step_n2.append(noise_2[ind])
        y_step_n3.append(noise_3[ind])
        ind+=1
    else:
        y_step_n1.append(noise_1[ind]+1)
        y_step_n2.append(noise_2[ind]+1)
        y_step_n3.append(noise_3[ind]+1)
        ind+=1

#Plotting functions with noise
plt.figure(figsize=(50, 8))
plt.subplot(1,3,1)
plt.axhline(y=0, c='black')
plt.axvline(x=0, c='black')
plt.grid()

plt.plot(x_step, y_step_n1, label='noise = 0.05')
plt.plot(x_step, y_step_n2, label='noise = 0.1')
plt.plot(x_step, y_step_n3, label='noise = 0.2')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('step functions with various noise')
plt.legend()

#First order derivative with noise
y_step_n1_d1 = np.diff(y_step_n1) / delta_x_step
y_step_n2_d1 = np.diff(y_step_n2) / delta_x_step
y_step_n3_d1 = np.diff(y_step_n3) / delta_x_step

plt.figure(figsize=(50, 8))
plt.subplot(1,3,2)
plt.axhline(y=0, c='black')
plt.axvline(x=0, c='black')
plt.grid()

plt.plot(x_step[0:-1], y_step_n1_d1, label = 'first order derivative with noise 0.05')
plt.plot(x_step[0:-1], y_step_n2_d1, label = 'first order derivative with noise 0.1')
plt.plot(x_step[0:-1], y_step_n3_d1, label = 'first order derivative with noise 0.2')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('First order derivatives of step function with noise')
plt.legend()

#Second order derivative with noise
y_step_n1_d2 = np.diff(y_step_n1_d1) / delta_x_step
y_step_n2_d2 = np.diff(y_step_n2_d1) / delta_x_step
y_step_n3_d2 = np.diff(y_step_n3_d1) / delta_x_step

plt.figure(figsize=(50, 8))
plt.subplot(1,3,3)
plt.axhline(y=0, c='black')
plt.axvline(x=0, c='black')
plt.grid()

plt.plot(x_step[0:-2], y_step_n1_d2, label = 'second order derivative with noise 0.05')
plt.plot(x_step[0:-2], y_step_n2_d2, label = 'second order derivative with noise 0.1')
plt.plot(x_step[0:-2], y_step_n3_d2, label = 'second order derivative with noise 0.2')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Second order derivatives of step function with noise')
plt.legend()

plt.show()
```

# Exercise 2

Write a code that

1. generate a step edge image like the one shown in the next cell
2. take the discrete first-order derivate and show the resulting image
3. take the discrete second-order derivate and show the resulting image

<img src="step edge.png" width=400 height=400 />

```python
#Anwser exersice 2.1
step_image = np.zeros((400,400))

for row in step_image:
    for index, _ in enumerate(row):  
        if index < len(row)/2:  
            row[index] = 1
        else:
            row[index] = 0

#Awnser exersice 2.2
delta_x_step_image = 1     #step between two pixels is 1 in this case
step_image_d1 = np.diff(step_image, axis=1) / delta_x_step_image       #derivative along the columns only with axis=1

#Anwser exersice 2.3
step_image_d2 = np.diff(step_image_d1, axis=1) / delta_x_step_image

#plotting
plt.figure(figsize=(15,30))

plt.subplot(1,3,1)
plt.title('image')
plt.imshow(step_image, cmap='Grays')

plt.subplot(1,3,2)
plt.title('first order derivative')
plt.imshow(step_image_d1, cmap='Grays')

plt.subplot(1,3,3)
plt.title('second order derivative')
plt.imshow(step_image_d2, cmap='Grays')

plt.show()
```

# Exercise 4

Write a code that

1. generate a ramp edge image like the one shown in the next cell
2. take the discrete first-order derivate and show the resulting image
3. take the discrete second-order derivate and show the resulting image

<img src="ramp edge.png" width=400 height=400 />

```python
#Anwser exersice 4.1
ramp_image = np.zeros((400,400))

for row in range(len(ramp_image)):
    for index in range(len(ramp_image[row])):

        if index < (len(ramp_image[row]) / 4):
            ramp_image[row][index] = 1  

        elif index >= (len(ramp_image[row]) / 4) and index < (len(ramp_image[row]) / 4 * 3):
            ramp_image[row][index] = 1 - (index - (len(ramp_image[row]) / 4)) / ((len(ramp_image[row]) / 2) - 1)  

        else:
            ramp_image[row][index] = 0  

#Awnser exersice 4.2
delta_x_ramp_image = 1     #step between two pixels is 1 in this case
ramp_image_d1 = np.diff(ramp_image, axis=1) / delta_x_ramp_image       #derivative along the columns only with axis=1

#Anwser exersice 4.3
ramp_image_d2 = np.diff(ramp_image_d1, axis=1) / delta_x_step_image

#plotting
plt.figure(figsize=(15,30))

plt.subplot(1,3,1)
plt.title('image')
plt.imshow(ramp_image, cmap='Grays')

plt.subplot(1,3,2)
plt.title('first order derivative')
plt.imshow(ramp_image_d1, cmap='Grays')

plt.subplot(1,3,3)
plt.title('second order derivative')
plt.imshow(ramp_image_d2, cmap='Grays')

plt.show()
```

```python
#Anwser exersice 6.1
roof_image = np.zeros((400, 400))

for row in range(len(roof_image)):
    for index in range(len(roof_image[row])):

        if index < ((len(roof_image[row]) / 10) * 4):
            roof_image[row][index] = 0

        elif index >= ((len(roof_image[row]) / 10) * 4) and index < ((len(roof_image[row]) / 10) * 6):
            roof_image[row][index] = 1 - abs(index - (len(roof_image[row]) / 2)) / ((len(roof_image[row]) / 5))

        else:
            roof_image[row][index] = 0

#Anwser exersice 6.2
delta_x_roof_image = 1
roof_image_d1 = np.diff(roof_image, axis=1) / delta_x_roof_image

#Anwser exersice 6.3
roof_image_d2 = np.diff(roof_image_d1, axis=1) / delta_x_roof_image

plt.figure(figsize=(15, 30))

plt.subplot(1, 3, 1)
plt.title('image')
plt.imshow(roof_image, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('first order derivative')
plt.imshow(roof_image_d1, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('second order derivative')
plt.imshow(roof_image_d2, cmap='gray')

plt.show()
```
