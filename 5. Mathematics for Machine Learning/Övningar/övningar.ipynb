```python
#image processing

import numpy as np
import cv2
import matplotlib.pyplot as plt

im = cv2.imread(r"C:\Users\victo\Pictures\lena.jpeg")
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)   #reading the image in grayscale

print(im.shape)

plt.imshow(im, cmap='Greys')
```

```python

kernel_x = np.array([-1, 0, 1])  #This is the filter that we manipulate each pixel with
masked_images = im.copy()

for i in range(1, im.shape[0] -1):  # We need to start at 1 in the range since we can not use the kernel on edge pixels. 
    for j in range(1, im.shape[1] -1):
        masked_images[i, j] = np.sum(np.multiply(im[i, j-1:j+2], kernel_x))     #Here we apply the kernel, or "mask" and asign the new pixel value to the new image

plt.imshow(masked_images, cmap='Greys') 
```
