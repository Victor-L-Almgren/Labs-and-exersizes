# Exersice 2 
1.1 Plot the following functions \
1.2 Calculate their derivatives using sympy \
1.3 Plot their derivatives on the same plot


- $f(x) = \frac{sin(x)}{x}$
- $f(x) = \frac{1}{1+e^{(-x)}}$

```python
# 2.1 Plot the functions

x = np.arange(-5, 5, 0.1)
y1 = (np.sin(x)) / x
y2 = 1 / (1 + np.exp(-x))

plt.figure(figsize = (15, 8))

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.axhline(y=0, c='black')
plt.axvline(x=0, c='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funtction 1')
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.axhline(y=0, c='black')
plt.axvline(x=0, c='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funtction 2')
plt.grid()

plt.show()
```

```python
# 2.2 Calculate their derivatives using sympy

x = sym.Symbol('x')

y1 = (sym.sin(x)) / x
y2 = 1 / (1 + sym.exp(-x))

der_y1 = y1.diff(x)
der_y2 = y2.diff(x)

print(f"Derivative of function 1:")
print(der_y1)
print(f"\nDerivative of function 2:")
print(der_y2)
```

```python
# 2.3 Plot the derivatives and the original functions overlaid

x = np.arange(-5, 5, 0.1)
y1 = (np.sin(x)) / x
y2 = 1 / (1 + np.exp(-x))

der_y1 = np.cos(x)/x - np.sin(x)/x**2
der_y2 = np.exp(-x)/(1 + np.exp(-x))**2

plt.figure(figsize = (15, 8))

plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.plot(x, der_y1)
plt.axhline(y=0, c='black')
plt.axvline(x=0, c='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funtction 1')
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.plot(x, der_y2)
plt.axhline(y=0, c='black')
plt.axvline(x=0, c='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funtction 2')
plt.grid()

plt.show()
```

# Exersice 3

-  Write a function that calculates numerical derivation without using symbolic computation
The function received $\Delta x$ and a function and the point around which the derivative will be calculated.

- Apply the function on the following functions to get calculate derivatives as some specific arbitrary points and compare the results with symbolic calculation (try for multiple value of $\Delta x$ to see the effect of the resolution of it


- $f(x) = 5x + 2$
- $f(x) = 6x^2$
- $f(x) = \frac{cos(2x+6)}{e^{6x}}$

```python
# function that calculates numerical derivation

def numerical_derivate(func, delta_x, point):
    return (func(point + delta_x) - func(point)) / delta_x 

# apply the numerical derivation function on the provided functions
def func1(x):
    return 5*x + 2

def func2(x):
    return 6*x**2

def func3(x):
    return np.cos(2*x + 6) / np.exp(6*x)
    
delta_x = 1e-15
point = 0

der_func1_num = numerical_derivate(func1, delta_x, point)
der_func2_num = numerical_derivate(func2, delta_x, point)
der_func3_num = numerical_derivate(func3, delta_x, point)

# Derivatives for the same points with symbolic calculation

x = sym.Symbol('x')

f1 = 5*x + 2
f2 = 6*x**2
f3 = sym.cos(2*x + 6) / sym.exp(6*x)

der_func1_sym = f1.diff(x)
der_func2_sym = f2.diff(x)
der_func3_sym = f3.diff(x)

val1 = der_func1_sym.subs(x, point).evalf()
val2 = der_func2_sym.subs(x, point).evalf()
val3 = der_func3_sym.subs(x, point).evalf()

# comparison between the two methods

print(f"Delta X = {delta_x} and the point is = {point} \n")
print(f"Numerical calculation of derivate for function 1 is {der_func1_num}")
print(f"Numerical calculation of derivate for function 2 is {der_func2_num}")
print(f"Numerical calculation of derivate for function 3 is {der_func3_num} \n")
print(f"Symbolic calculation of derivate for function 1 is {val1}")
print(f"Symbolic calculation of derivate for function 2 is {val2}")
print(f"Symbolic calculation of derivate for function 3 is {val3} \n")
```
